# Iris Dataset Using Scikit-Learn

## Machine Learning on the Iris Dataset

Is a supervised learning problem: predict the species of an iris depending on the measurements.

Each value we are predicting is called the response: also known as target, outcome, label

**Classification** is supervised learning in which the response in categorical

**Regression** is supervised learning in which the response in ordered and continuous

## Splitting Input Columns and Output Columns

```python
data: ndarray = df.values
```

```python
# Seperate the features and the target
# feature values
features_data: ndarray = data[:, 0:4]
# target values
target_data: ndarray = data[:, 4]
```

## Splitting Data into Training and Testing

```python
from sklearn.model_selection import train_test_split

# 80% training, 20% testing
features_data_train, features_data_test, target_data_train, target_data_test = train_test_split(features_data, target_data, test_size=0.2)
```

## Model 1: Support Vector Machine Algorithm

```python
from sklearn.svm import SVC

# model fitting
model_svc = SVC()
model_svc.fit(features_data_train, target_data_train)
```

```python
# model testing
prediction_svc: ndarray = model_svc.predict(features_data_test)
```

```python
from sklearn.metrics import accuracy_score

# Calculate the accuracy
print(accuracy_score(target_data_test, prediction_svc)*100, "\b%")
```

## Model 2: Linear Regression

```python
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression()
model_lr.fit(features_data_train, target_data_train)
```

```python
prediction_lr: ndarray = model_lr.predict(features_data_test)
```

```python
print(accuracy_score(target_data_test, prediction_lr)*100, "\b%")
```

## Model 3: Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

model_dt = DecisionTreeClassifier()
model_dt.fit(features_data_train, target_data_train)
```

```python
prediction_dt: ndarray = model_dt.predict(features_data_test)
```

```python
print(accuracy_score(target_data_test, prediction_dt)*100, "\b%")
```

## Model 4: Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model_ll = LogisticRegression()
model_ll.fit(features_data_train, target_data_train)
```

```python
prediction_ll: ndarray = model_ll.predict(features_data_test)
```

```python
print(accuracy_score(target_data_test, prediction_ll)*100, "\b%")
```

### Classification Report

Check the precision, recall, f1-score

```python
# A detailed classification report
from sklearn.metrics import classification_report

print(classification_report(target_data_test, prediction_lr))
```
